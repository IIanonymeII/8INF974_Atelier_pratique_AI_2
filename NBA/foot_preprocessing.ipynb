```python
#   X   # away_current_pos Classement actuelle de l’équipe "away"
#   X   # away_last_yr_pos Classement de l’équipe "away" l’année précédente
#===V===# away_prev_game_perf Performance de l’équipe "away" lors du dernier match
#===V===# away_team_av_points Points marqués en moyenne par l’équipe "away"
#===V===# away_team_av_points_conceded Points concédés en moyenne par l’équipe "away"
#===V===# away_team_away_form Forme de l’équipe "away" à l’extérieur
#===V===# away_team_form Forme de l’équipe "away"
#===V===# away_team_rest_time Temps de repos de l’équipe "away"
#===V===# away_win_percentage Pourcentage de victoires de l’équipe "away"
#   X   # h2h_form Forme en tête-à-tête entre les équipes
#   X   # home_current_pos Classement actuelle de l’équipe "home"
#   X   # home_last_yr_pos Classement de l’équipe "home" l’année précédente
#===V===# home_prev_game_perf Performance de l’équipe "home" lors du dernier match
#===V===# home_team_av_points Points marqués en moyenne par l’équipe "home"
#===V===# home_team_av_points_conceded Points concédés en moyenne par l’équipe "home"
#===V===# home_team_form Forme de l’équipe "home"
#===V===# home_team_home_form Forme de l’équipe "home" à domicile
#===V===# home_team_rest_time Temps de repos de l’équipe "home"
#===V===# home_win_percentage Pourcentage de victoires de l’équipe "home"
#   X   # match_importance Importance du match (ex. finale, demi-finale)
#===V===# Date Jour du match
#===V===# away_team Nom de l’équipe "away"
#===V===# home_team Nom de l’équipe "home"
#===V===# month Mois du match
#===V===# year Année du match
#===V===# result
```

```python
import numpy as np 
import pandas as pd
```

```python
games = pd.read_csv("../Data/Foot/Foot_Seasons_Raw.csv")

games_columns_keep = ['Div', 'Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR']
games = games[games_columns_keep]

games['game_date_est'] = pd.to_datetime(games['Date'])

games["year"] = games["game_date_est"].dt.year
games["month"] = games["game_date_est"].dt.month
games["date"] = games["game_date_est"].dt.day

print(" ==== STAT ====")
print(games["FTR"].value_counts())
# case were just WIN or LOSE
# NOT NULL
games = games[games["FTR"]!='D']


games.head(10)
```

```python
def map_result(result):
    if result == 'A':
        return 0
    elif result == 'H':
        return 1
    else:
        return 2

games['result'] = games['FTR'].apply(map_result)
games.head(10)
```

```python
def inverse_result(result):
    if result == 0:
        return 1
    elif result == 1:
        return 0
    else:
        return 2


transformed_df = pd.DataFrame(columns=['game_date_est','team', 'result','pts','pts_conceded'])

transformed_df = pd.concat([transformed_df, games[['game_date_est','AwayTeam', 'result','FTAG','FTHG']].rename(columns={'game_date_est':'game_date_est', 'AwayTeam': 'team', 'result': 'result','FTAG':'pts','FTHG':'pts_conceded'})])

# Inverser les résultats
transformed_df['result'] = transformed_df['result'].apply(inverse_result)

transformed_df = pd.concat([transformed_df, games[['game_date_est','HomeTeam', 'result','FTHG','FTAG']].rename(columns={'game_date_est':'game_date_est', 'HomeTeam': 'team', 'result': 'result','FTHG':'pts','FTAG':'pts_conceded'})])



transformed_df = transformed_df.sort_values(by='game_date_est')


# ====== PREVIOUS RESULT ========

transformed_df['previous_result']   = transformed_df.groupby(["team"])["result"].shift(1)

# ====== POURCENTAGE WIN =====

# Calculer la somme cumulée des résultats précédents pour chaque équipe
transformed_df['cumulative_sum'] = transformed_df.groupby(["team"])["result"].cumsum()

# Calculer le nombre total de résultats pour chaque équipe
transformed_df['total_results'] = transformed_df.groupby('team')['result'].cumcount() + 1

# Calculer le pourcentage de victoires pour chaque équipe
transformed_df['win_percentage'] = transformed_df['cumulative_sum'] / transformed_df['total_results']

# Supprimer les colonnes temporaires créées
transformed_df.drop(['cumulative_sum', 'total_results'], axis=1, inplace=True)



# ====== FORM ========
transformed_df['previous_result_2'] = transformed_df.groupby(["team"])["result"].shift(2)
transformed_df['previous_result_3'] = transformed_df.groupby(["team"])["result"].shift(3)
transformed_df['previous_result_4'] = transformed_df.groupby(["team"])["result"].shift(4)
transformed_df['previous_result_5'] = transformed_df.groupby(["team"])["result"].shift(5)

transformed_df["form_5_previous_match"] = transformed_df[['previous_result_5', 'previous_result_4', 'previous_result_3', 'previous_result_2', 'previous_result']].sum(axis=1, skipna=False)
transformed_df  = transformed_df.drop(columns=[ 'previous_result_5', 'previous_result_4', 'previous_result_3', 'previous_result_2'])

 # ===== PAUSE =======
transformed_df['game_date_est'] = pd.to_datetime(transformed_df['game_date_est'])

# Calculer la différence des dates pour chaque équipe
transformed_df['date_difference'] = transformed_df.groupby(["team"])["game_date_est"].diff()

# Plafonner la différence à 7 jours
transformed_df['date_difference'] = transformed_df['date_difference'].clip(upper=pd.Timedelta(days=7))

# Convertir la différence plafonnée en entiers (nombre de jours)
transformed_df['date_difference'] = transformed_df['date_difference'].dt.days
transformed_df['game_date_est'] = transformed_df['game_date_est'].astype(str)


# ====== Moyenne pts =====
transformed_df['pts_shift'] = transformed_df.groupby('team')['pts'].shift(1)
transformed_df['pts_conceded_shift'] = transformed_df.groupby('team')['pts_conceded'].shift(1)
     
transformed_df['moyenne_pts_25'] = transformed_df.groupby('team')['pts_shift'].transform(lambda x: x.rolling(window=25, min_periods=1).mean())
transformed_df['moyenne_pts_conceded_25'] = transformed_df.groupby('team')['pts_conceded_shift'].transform(lambda x: x.rolling(window=25, min_periods=1).mean())
transformed_df  = transformed_df.drop(columns=[ 'pts_shift','pts_conceded_shift'])


transformed_df["win"] =  transformed_df.groupby(["team"])["result"].cumsum()
transformed_df["win"] =  transformed_df.groupby(["team"])["win"].shift(1)

transformed_df["nbr_match"] =  transformed_df.groupby(["team"])["result"].cumcount()
transformed_df["lose"] = transformed_df["nbr_match"] - transformed_df["win"]
transformed_df  = transformed_df.drop(columns=[ "nbr_match"])


transformed_df[transformed_df["team"] == "West Brom"].head(10)
```

```python
games["game_date_est"] = games['game_date_est'].astype(str)
previous_info_merge_away = pd.merge(games, transformed_df, left_on=['game_date_est', 'AwayTeam'], right_on=['game_date_est', 'team'], how='inner')
previous_info_merge_away = previous_info_merge_away.rename(columns={
    "previous_result" : "away_prev_game_perf",
    "form_5_previous_match" : "away_team_form",
    "date_difference" : "away_team_rest_time",
    "moyenne_pts_25" : "away_team_av_points",
    "moyenne_pts_conceded_25" : "away_team_av_points_conceded",
    "win_percentage" : "away_win_percentage",
    "win":"away_win",
    "lose":"away_lose",
    

})
# previous_info_merge_away.columns

previous_info_merge_both = pd.merge(previous_info_merge_away, transformed_df, left_on=['game_date_est', 'HomeTeam'], right_on=['game_date_est', 'team'], how='inner')
previous_info_merge_both = previous_info_merge_both.rename(columns={
    "previous_result" : "home_prev_game_perf",
    "form_5_previous_match" : "home_team_form",
    "date_difference" : "home_team_rest_time",
    "moyenne_pts_25" : "home_team_av_points",
    "moyenne_pts_conceded_25" : "home_team_av_points_conceded",    
    "win_percentage" : "home_win_percentage",
    "win":"home_win",
    "lose":"home_lose",
})
# previous_info_merge_both.columns
games_transi = previous_info_merge_both.drop(columns=[
                                                        # "game_date_est_x","game_date_est_y",
                                                        "result_y","result_x",
                                                        "team_x","team_y",
                                                        "pts_x","pts_y",
                                                        "pts_conceded_x","pts_conceded_y",
                                            #    "win_percentage_x","win_percentage_y",
                                               ])

print(f"avant : {len(games)}")
print(f"apres : {len(games_transi)}")
games_transi.columns
```

```python
game_copy = games_transi.copy()

game_copy['previous_result_1_away'] = game_copy.groupby(["AwayTeam"])["result"].shift(1).replace({1: 0, 0: 1})
game_copy['previous_result_2_away'] = game_copy.groupby(["AwayTeam"])["result"].shift(2).replace({1: 0, 0: 1})
game_copy['previous_result_3_away'] = game_copy.groupby(["AwayTeam"])["result"].shift(3).replace({1: 0, 0: 1})
game_copy['previous_result_4_away'] = game_copy.groupby(["AwayTeam"])["result"].shift(4).replace({1: 0, 0: 1})
game_copy['previous_result_5_away'] = game_copy.groupby(["AwayTeam"])["result"].shift(5).replace({1: 0, 0: 1})
game_copy['previous_result_6_away'] = game_copy.groupby(["AwayTeam"])["result"].shift(6).replace({1: 0, 0: 1})
game_copy['previous_result_7_away'] = game_copy.groupby(["AwayTeam"])["result"].shift(7).replace({1: 0, 0: 1})
game_copy['previous_result_8_away'] = game_copy.groupby(["AwayTeam"])["result"].shift(8).replace({1: 0, 0: 1})
game_copy['previous_result_9_away'] = game_copy.groupby(["AwayTeam"])["result"].shift(9).replace({1: 0, 0: 1})
game_copy['previous_result_10_away'] = game_copy.groupby(["AwayTeam"])["result"].shift(10).replace({1: 0, 0: 1})


game_copy["away_team_away_form"] = game_copy[['previous_result_10_away','previous_result_9_away','previous_result_8_away','previous_result_7_away','previous_result_6_away','previous_result_5_away', 'previous_result_4_away', 'previous_result_3_away', 'previous_result_2_away', 'previous_result_1_away']].sum(axis=1, skipna=False)
game_copy  = game_copy.drop(columns=['previous_result_10_away','previous_result_9_away','previous_result_8_away','previous_result_7_away','previous_result_6_away','previous_result_5_away', 'previous_result_4_away', 'previous_result_3_away', 'previous_result_2_away', 'previous_result_1_away'])

game_copy['previous_result_1_home'] = game_copy.groupby(["HomeTeam"])["result"].shift(1)
game_copy['previous_result_2_home'] = game_copy.groupby(["HomeTeam"])["result"].shift(2)
game_copy['previous_result_3_home'] = game_copy.groupby(["HomeTeam"])["result"].shift(3)
game_copy['previous_result_4_home'] = game_copy.groupby(["HomeTeam"])["result"].shift(4)
game_copy['previous_result_5_home'] = game_copy.groupby(["HomeTeam"])["result"].shift(5)
game_copy['previous_result_6_home'] = game_copy.groupby(["HomeTeam"])["result"].shift(5)
game_copy['previous_result_7_home'] = game_copy.groupby(["HomeTeam"])["result"].shift(7)
game_copy['previous_result_8_home'] = game_copy.groupby(["HomeTeam"])["result"].shift(8)
game_copy['previous_result_9_home'] = game_copy.groupby(["HomeTeam"])["result"].shift(9)
game_copy['previous_result_10_home'] = game_copy.groupby(["HomeTeam"])["result"].shift(10)

game_copy["home_team_home_form"] = game_copy[['previous_result_10_home','previous_result_9_home','previous_result_8_home','previous_result_7_home','previous_result_6_home','previous_result_5_home', 'previous_result_4_home', 'previous_result_3_home', 'previous_result_2_home', 'previous_result_1_home']].sum(axis=1, skipna=False)
game_copy  = game_copy.drop(columns=['previous_result_10_home','previous_result_9_home','previous_result_8_home','previous_result_7_home','previous_result_6_home','previous_result_5_home', 'previous_result_4_home', 'previous_result_3_home', 'previous_result_2_home', 'previous_result_1_home'])

game_copy.columns


# ========= HOME =========
game_copy["Homewon"] = game_copy.groupby(["HomeTeam"])["result"].cumsum()
game_copy["Homewon"] = game_copy.groupby(["HomeTeam"])["result"].shift(1)

game_copy["nbr_match"] =  game_copy.groupby(["HomeTeam"])["result"].cumcount()
game_copy["Homelost"] = game_copy["nbr_match"] - game_copy["Homewon"]
game_copy  = game_copy.drop(columns=[ "nbr_match"])

# ========= AWAY =========
game_copy["Roadwon"] = game_copy.groupby(["AwayTeam"])["result"].cumsum()
game_copy["Roadwon"] = game_copy.groupby(["AwayTeam"])["result"].shift(1)

game_copy["nbr_match"] =  game_copy.groupby(["AwayTeam"])["result"].cumcount()
game_copy["Roadlost"] = game_copy["nbr_match"] - game_copy["Roadwon"]
game_copy  = game_copy.drop(columns=[ "nbr_match"])


print(f"avant : {len(games_transi)}")
print(f"apres : {len(game_copy)}")
```

```python
final_game = game_copy.copy()
final_game = final_game.drop(columns=["Date"])
final_game[final_game["HomeTeam"] == "West Brom"][["game_date_est","AwayTeam","HomeTeam","result","away_team_away_form","home_team_home_form"]].head(10)
final_game.columns
```

```python
final_game["away_last_yr_pos"] = 0

final_game["h2h_form"] = 0
final_game["home_current_pos"] = 0
final_game["home_last_yr_pos"] = 0
final_game["match_importance"] = 0


columns_to_have = [
                   "away_team_form",
                   "away_win_percentage",
                   "away_win",
                   "away_lose",
                   "Roadwon",
                   "Roadlost",

                   "home_team_form",
                   "home_win_percentage",
                   "home_win",
                   "home_lose",
                   "Homewon",
                   "Homelost",

                   "result",
                   "AwayTeam",
                   "HomeTeam",
                   ]


convert_dict = {
                "away_L10"  : "away_team_form"      , 
                "away_Pct"  : "away_win_percentage" ,  
                "away_Won"  : "away_win"            ,
                "away_Lost" : "away_lose"           ,

                "home_L10"  : "home_team_form"      , 
                "home_Pct"  : "home_win_percentage" , 
                "home_Won"  : "home_win"            ,
                "home_lose" : "home_lose"           ,
}
final_game_csv = final_game.rename(columns=convert_dict)
final_game_csv = final_game_csv[columns_to_have]
final_game_csv
```

```python
final_game_csv = final_game_csv.dropna(subset=columns_to_have)
final_game_csv.info()
```
