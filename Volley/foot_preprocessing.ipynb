```python
import numpy as np 
import pandas as pd
```

```python
games = pd.read_csv("../Data/Foot/Foot_Seasons_Raw.csv")

games_columns_keep = ['Div', 'Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR']
games = games[games_columns_keep]

games['Date'] = pd.to_datetime(games['Date'])

games["year"] = games["Date"].dt.year
games["month"] = games["Date"].dt.month
games["date"] = games["Date"].dt.day

games = 

games.head(10)
```

```python
def map_result(result):
    if result == 'A':
        return 0
    elif result == 'H':
        return 1
    else:
        return 2

games['result'] = games['FTR'].apply(map_result)
games.head(10)
```

```python
def inverse_result(result):
    if result == 0:
        return 1
    elif result == 1:
        return 0
    else:
        return 2


transformed_df = pd.DataFrame(columns=['date','team', 'result','pts','pts_conceded'])

transformed_df = pd.concat([transformed_df, games[['Date','AwayTeam', 'result','FTAG','FTHG']].rename(columns={'Date':'date', 'AwayTeam': 'team', 'result': 'result','FTAG':'pts','FTHG':'pts_conceded'})])

# Inverser les résultats
transformed_df['result'] = transformed_df['result'].apply(inverse_result)


transformed_df = pd.concat([transformed_df, games[['Date','HomeTeam', 'result','FTHG','FTAG']].rename(columns={'Date':'date', 'HomeTeam': 'team', 'result': 'result','FTHG':'pts','FTAG':'pts_conceded'})])



transformed_df = transformed_df.sort_values(by='date')


# ====== PREVIOUS RESULT ========

transformed_df['previous_result']   = transformed_df.groupby(["team"])["result"].shift(1)



# ====== FORM ========
transformed_df['previous_result_2'] = transformed_df.groupby(["team"])["result"].shift(2)
transformed_df['previous_result_3'] = transformed_df.groupby(["team"])["result"].shift(3)
transformed_df['previous_result_4'] = transformed_df.groupby(["team"])["result"].shift(4)
transformed_df['previous_result_5'] = transformed_df.groupby(["team"])["result"].shift(5)

transformed_df["form_5_previous_match"] = transformed_df[['previous_result_5', 'previous_result_4', 'previous_result_3', 'previous_result_2', 'previous_result']].sum(axis=1, skipna=False)
transformed_df  = transformed_df.drop(columns=[ 'previous_result_5', 'previous_result_4', 'previous_result_3', 'previous_result_2'])

transformed_df[transformed_df["team"] == "West Brom"].head(10)
```
