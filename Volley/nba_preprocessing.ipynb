```python
import numpy as np 
import pandas as pd
```

```python
games = pd.read_csv("../Data/NBA/games.csv")

games_columns_keep = [ "GAME_ID",
                 "GAME_DATE_EST", # Date
                 "TEAM_ID_away", # name away
                 "TEAM_ID_home", # name home
                 "GAME_STATUS_TEXT", # final, demi,...
                 "PTS_home", "PTS_away",
                 "HOME_TEAM_WINS" #result
                 ]
games['GAME_DATE_EST'] = pd.to_datetime(games['GAME_DATE_EST'])
games = games[games_columns_keep]
games["year"] = games["GAME_DATE_EST"].dt.year
games["month"] = games["GAME_DATE_EST"].dt.month
games["date"] = games["GAME_DATE_EST"].dt.day

games.head()
# games.columns
```

```python
teams = pd.read_csv("../Data/NBA/teams.csv")

teams_columns_keep = ["TEAM_ID",
                    "CITY"]

teams = teams[teams_columns_keep]
teams.head()
```

```python
home_games = pd.merge(games, teams, left_on = "TEAM_ID_home", right_on = "TEAM_ID", how = "inner")


home_games = home_games.rename(columns={
    "CITY" : "home_team",
})

home_games.head()
```

```python
both_games = pd.merge(home_games, teams, left_on= "TEAM_ID_away", right_on="TEAM_ID", how="inner")

both_games = both_games.rename(columns={
    "CITY" : "away_team",
    "HOME_TEAM_WINS" : "result"
})

# both_games = both_games.drop(columns=[
#     "TEAM_ID_home", "TEAM_ID_away", "TEAM_ID_x", "TEAM_ID_y"
# ])
both_games.head()
```

```python
games = both_games.copy()
games.columns = games.columns.str.lower()
games["total_points"] = games["pts_home"] + games["pts_away"]
games.head()
```

```python
ranking = pd.read_csv("../Data/NBA/ranking.csv")

ranking_columns_keep = ["TEAM_ID",
                        "STANDINGSDATE",
                        "G", # Nombre total de matchs joués
                        "W", # Nombre de matchs gagnés.
                        "L", # Nombre de matchs perdus.
                        "W_PCT", # Pourcentage de victoires
                        "HOME_RECORD", # Home gagné - home perdu
                        "ROAD_RECORD", # away gagné - away perdu
                        ]

ranking = ranking[ranking_columns_keep]
ranking.head()
```

```python
games['game_date_est'] = games['game_date_est'].astype(str)

# ranking["STANDINGSDATE"] == str(games["game_date_est"]) and ranking["TEAM_ID"] == str(games["TEAM_ID_away"])
merged_ranking_away = pd.merge(ranking, games, left_on=['STANDINGSDATE', 'TEAM_ID'], right_on=['game_date_est', 'team_id_away'], how='inner')
merged_ranking_away = merged_ranking_away.rename(columns={
    "W_PCT" : "away_win_percentage",
    "W" : "away_win_this_season",
    "L" : "away_lose_this_season",
    "HOME_RECORD" : "away_record_home",
    "ROAD_RECORD" : "away_record_away"
})


merged_ranking_both = pd.merge(ranking, merged_ranking_away, left_on=['STANDINGSDATE', 'TEAM_ID'], right_on=['STANDINGSDATE', 'team_id_home'], how='inner')

merged_ranking_both = merged_ranking_both.rename(columns={
    "W_PCT" : "home_win_percentage",
    "W" : "home_win_this_season",
    "L" : "home_lose_this_season",
    "HOME_RECORD" : "home_record_home",
    "ROAD_RECORD" : "home_record_away"
})


# games = merged_ranking_both.copy()
merged_ranking_both.columns
```

```python
games = merged_ranking_both.drop(columns=[
    "TEAM_ID_x", "TEAM_ID_y", 
    "STANDINGSDATE",
    "G_x", "G_y", 
    "team_id_x","team_id_y"
])
games.columns

```

```python
# Trier les données par 'game_date_est' pour chaque équipe
games_df_sorted = games.sort_values(by='game_date_est').copy()
# games_df_sorted.head()
# # Créer une colonne 'last_game_result' pour stocker le résultat du dernier match de chaque équipe
def new_df_previous_result(df_copy):
    transformed_df = pd.DataFrame(columns=['game_date_est','team', 'result','pts','pts_conceded'])

    transformed_df = pd.concat([transformed_df, df_copy[['game_date_est','team_id_away', 'result','pts_away','pts_home']].rename(columns={'team_id_away': 'team', 'result': 'result','pts_away':'pts','pts_home':'pts_conceded'})])
    
    # Inverser les résultats
    transformed_df['result'] = transformed_df.apply(lambda row: 0 if row['result'] == 1 else 1, axis=1)

    transformed_df = pd.concat([transformed_df, df_copy[['game_date_est','team_id_home', 'result','pts_home','pts_away']].rename(columns={'team_id_home': 'team', 'result': 'result','pts_home':'pts','pts_away':'pts_conceded'})])

    transformed_df = transformed_df.sort_values(by='game_date_est')

    transformed_df['previous_result']   = transformed_df.groupby(["team"])["result"].shift(1)

    # ====== FORM ========
    transformed_df['previous_result_2'] = transformed_df.groupby(["team"])["result"].shift(2)
    transformed_df['previous_result_3'] = transformed_df.groupby(["team"])["result"].shift(3)
    transformed_df['previous_result_4'] = transformed_df.groupby(["team"])["result"].shift(4)
    transformed_df['previous_result_5'] = transformed_df.groupby(["team"])["result"].shift(5)
    
    transformed_df["form_5_previous_match"] = transformed_df[['previous_result_5', 'previous_result_4', 'previous_result_3', 'previous_result_2', 'previous_result']].sum(axis=1, skipna=False)
    transformed_df  = transformed_df.drop(columns=[ 'previous_result_5', 'previous_result_4', 'previous_result_3', 'previous_result_2'])

    # ===== PAUSE =======
    transformed_df['game_date_est'] = pd.to_datetime(transformed_df['game_date_est'])

    # Calculer la différence des dates pour chaque équipe
    transformed_df['date_difference'] = transformed_df.groupby(["team"])["game_date_est"].diff()

    # Plafonner la différence à 7 jours
    transformed_df['date_difference'] = transformed_df['date_difference'].clip(upper=pd.Timedelta(days=7))

    # Convertir la différence plafonnée en entiers (nombre de jours)
    transformed_df['date_difference'] = transformed_df['date_difference'].dt.days
    transformed_df['game_date_est'] = transformed_df['game_date_est'].astype(str)

    # ====== Moyenne pts =====
    transformed_df['pts_shift'] = transformed_df.groupby('team')['pts'].shift(1)
    transformed_df['pts_conceded_shift'] = transformed_df.groupby('team')['pts_conceded'].shift(1)

    transformed_df['moyenne_pts_25'] = transformed_df.groupby('team')['pts_shift'].transform(lambda x: x.rolling(window=25, min_periods=1).mean())
    transformed_df['moyenne_pts_conceded_25'] = transformed_df.groupby('team')['pts_conceded_shift'].transform(lambda x: x.rolling(window=25, min_periods=1).mean())
    transformed_df  = transformed_df.drop(columns=[ 'pts_shift','pts_conceded_shift'])



    

    return transformed_df
    

team_result_df = new_df_previous_result(df_copy=games_df_sorted)


team_result_df[(team_result_df["team"] == 1610612742)].head(15)
```

```python
previous_info_merge_away = pd.merge(games, team_result_df, left_on=['game_date_est', 'team_id_away'], right_on=['game_date_est', 'team'], how='inner')
previous_info_merge_away = previous_info_merge_away.rename(columns={
    "previous_result" : "away_prev_game_perf",
    "form_5_previous_match" : "away_team_form",
    "date_difference" : "away_team_rest_time",
    "moyenne_pts_25" : "away_team_av_points",
    "moyenne_pts_conceded_25" : "away_team_av_points_conceded",
    

})
previous_info_merge_away.columns

previous_info_merge_both = pd.merge(previous_info_merge_away, team_result_df, left_on=['game_date_est', 'team_id_home'], right_on=['game_date_est', 'team'], how='inner')
previous_info_merge_both = previous_info_merge_both.rename(columns={
    "previous_result" : "home_prev_game_perf",
    "form_5_previous_match" : "home_team_form",
    "date_difference" : "home_team_rest_time",
    "moyenne_pts_25" : "home_team_av_points",
    "moyenne_pts_conceded_25" : "home_team_av_points_conceded",
})

games = previous_info_merge_both.drop(columns=[ "result_y","result_x","team_x","pts_x","pts_conceded_x","team_y","pts_y","pts_conceded_y"])
games.columns
```

```python

# games["game_status_text"].head(10)
game_copy = games.copy()

game_copy['previous_result_1_away'] = game_copy.groupby(["team_id_away"])["result"].shift(1).replace({1: 0, 0: 1})
game_copy['previous_result_2_away'] = game_copy.groupby(["team_id_away"])["result"].shift(2).replace({1: 0, 0: 1})
game_copy['previous_result_3_away'] = game_copy.groupby(["team_id_away"])["result"].shift(3).replace({1: 0, 0: 1})
game_copy['previous_result_4_away'] = game_copy.groupby(["team_id_away"])["result"].shift(4).replace({1: 0, 0: 1})
game_copy['previous_result_5_away'] = game_copy.groupby(["team_id_away"])["result"].shift(5).replace({1: 0, 0: 1})

game_copy["away_team_away_form"] = game_copy[['previous_result_5_away', 'previous_result_4_away', 'previous_result_3_away', 'previous_result_2_away', 'previous_result_1_away']].sum(axis=1, skipna=False)
game_copy  = game_copy.drop(columns=['previous_result_5_away', 'previous_result_4_away', 'previous_result_3_away', 'previous_result_2_away', 'previous_result_1_away'])

game_copy['previous_result_1_home'] = game_copy.groupby(["team_id_home"])["result"].shift(1)
game_copy['previous_result_2_home'] = game_copy.groupby(["team_id_home"])["result"].shift(2)
game_copy['previous_result_3_home'] = game_copy.groupby(["team_id_home"])["result"].shift(3)
game_copy['previous_result_4_home'] = game_copy.groupby(["team_id_home"])["result"].shift(4)
game_copy['previous_result_5_home'] = game_copy.groupby(["team_id_home"])["result"].shift(5)

game_copy["home_team_home_form"] = game_copy[['previous_result_5_home', 'previous_result_4_home', 'previous_result_3_home', 'previous_result_2_home', 'previous_result_1_home']].sum(axis=1, skipna=False)
game_copy  = game_copy.drop(columns=['previous_result_5_home', 'previous_result_4_home', 'previous_result_3_home', 'previous_result_2_home', 'previous_result_1_home'])


game_copy.columns
```

```python
final_game = game_copy.copy()
final_game[final_game["team_id_away"] == 1610612759 ][["team_id_away","team_id_home","result","away_team_away_form","home_team_home_form"]].head(10)
```

```python
final_game["away_last_yr_pos"] = 0

final_game["h2h_form"] = 0
final_game["home_current_pos"] = 0
final_game["home_last_yr_pos"] = 0
final_game["match_importance"] = 0


columns_to_have = [
                   "away_last_yr_pos",
                   "away_prev_game_perf",
                   "away_team_av_points",
                   "away_team_av_points_conceded",
                   "away_team_away_form",
                   "away_team_form",
                   "away_team_rest_time",
                   "away_win_percentage",
                   "h2h_form",
                   "home_current_pos",
                   "home_last_yr_pos",
                   "home_prev_game_perf",
                   "home_team_av_points",
                   "home_team_av_points_conceded",
                   "home_team_form",
                   "home_team_home_form",
                   "home_team_rest_time",
                   "home_win_percentage",
                   "match_importance",
                   "result",
                   "Date",
                   "away_team",
                   "home_team",
                   "month",
                   "year"]


convert_dict = {
                # None                             :"away_last_yr_pos"            , 
                "away_prev_game_perf"            :"away_prev_game_perf"         , 
                "away_team_av_points"            :"away_team_av_points"         , 
                "away_team_av_points_conceded"   :"away_team_av_points_conceded", 
                "away_team_away_form"            :"away_team_away_form"         , 
                "away_team_form"                 :"away_team_form"              , 
                "away_team_rest_time"            :"away_team_rest_time"         , 
                "away_win_percentage"            :"away_win_percentage"         , 
                # None                             :"h2h_form"                    , 
                # None                             :"home_current_pos"            , 
                # None                             :"home_last_yr_pos"            , 
                "home_prev_game_perf"            :"home_prev_game_perf"         , 
                "home_team_av_points"            :"home_team_av_points"         , 
                "home_team_av_points_conceded"   :"home_team_av_points_conceded", 
                "home_team_form"                 :"home_team_form"              , 
                "home_team_home_form"            :"home_team_home_form"         , 
                "home_team_rest_time"            :"home_team_rest_time"         , 
                "home_win_percentage"            :"home_win_percentage"         , 
                # None                             :"match_importance"            , 
                "result"                         :"result"                      , 
                "date"                           :"Date"                        , 
                "away_team"                      :"away_team"                   , 
                "home_team"                      :"home_team"                   , 
                "month"                          :"month"                       , 
                "year"                           :"year"                         
}
final_game_csv = final_game.rename(columns=convert_dict)
final_game_csv = final_game_csv[columns_to_have]
final_game_csv
```
